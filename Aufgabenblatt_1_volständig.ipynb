Empirische Kovarianz: $$\widehat s_{xy}=\frac{1}{n-1}\sum_{i=1}^n(x_i−x ̅)(y_i−y ̅)$$



```python
from math import sqrt
age = [30, 24, 21, 18, 25, 20, 33, 25, 33, 20, 24, 22, 28, 32, 30, 32, 24, 20, 
       18, 21, 17, 22, 20, 30, 23, 32, 18, 32, 30, 28, 19, 26, 32, 23, 32, 30, 
       21, 29, 30, 20, 22, 33, 28, 29, 32, 31, 18, 32, 28, 24, 24, 31, 31, 17, 
       17, 30, 23, 19, 24, 34, 20, 31, 21, 31, 34, 23, 21, 27, 34, 24, 26, 22, 
       34, 28, 18, 19, 30, 24, 28, 27, 21, 26, 18, 29, 26, 29, 29, 31, 22, 31, 
       22, 34, 34, 33, 20, 27, 25, 25, 19, 18]
mark = [1.0 , 3.1, 4.6, 1.6, 2.5, 2.5, 4.9, 1.9, 3.7, 1.0 , 1.9, 4. , 2.8,
       4.3, 2.2, 3.1, 1.6, 4.3, 3.7, 1.6, 2.5, 1.0 , 1.3, 4. , 3.7, 1.3,
       3.7, 4.3, 2.5, 4. , 4.6, 3.4, 4.3, 3.4, 1.9, 1.0 , 3.7, 4. , 1.9,
       4.3, 2.2, 1.9, 1.9, 1.6, 1.3, 2.5, 1.9, 1.0 , 2.5, 3.7, 1.9, 1.0 ,
       3.4, 1.0 , 3.7, 3.4, 4.6, 3.1, 3.7, 4.9, 2.5, 2.2, 3.4, 3.1, 4.3,
       1.0 , 4.6, 1.9, 2.8, 1.3, 1.0 , 4. , 3.1, 3.4, 2.8, 1.0 , 1.9, 1.0 ,
       1.3, 3.1, 2.5, 1.3, 1.3, 2.8, 2.8, 1.3, 1.3, 3.4, 4.3, 4.9, 2.2,
       4.3, 4.6, 1.9, 1.3, 4. , 3.1, 3.4, 3.1, 3.4]

# arithmetische Mittelwert
M_age = sum(age)/len(age)
print(M_age) 
M_mark = sum(mark)/len(mark)
print(M_mark) 
# Stichprobe
Anzahl = len(mark)-1
print(Anzahl)

#Empirische Kovarianz
Sub_age = [i - M_age for i in age]
Sub_mark = [i - M_mark for i in mark]
Zähler = sum([Sub_age[i]*Sub_mark[i] for i in range(len(Sub_mark))]) 
kov = Zähler/Anzahl  
print("Empirische Kovarianz: = ",kov)

   


```

Empirische Standardabweichung: $$s_x=\sqrt{\frac{1}{n−1}\sum_{i=1}^{n}(x_i−x ̅)^2}$$

```python
#Empirische Standardabweichung

std_abw_age = sqrt(sum([Sub_age[i]**2.0 for i in range(len(Sub_age))])/Anzahl)

std_abw_mark = sqrt(sum([Sub_mark[i]**2.0 for i in range(len(Sub_mark))])/Anzahl)

print("Standard_Abweichung_age = ",std_abw_age, "Standard_Abweichung_mark = ", std_abw_mark)
```

Pearson-Korrelationskoeffizient: $$R_{xy}=\frac{\widehat s_{xy}}{s_xs_y}$$

```python
#Pearson-Korrelationskoeffizient
P_Korr = kov/(std_abw_age*std_abw_mark)
print("Pearson-Korrelationskoeffizient = ", P_Korr)

```
